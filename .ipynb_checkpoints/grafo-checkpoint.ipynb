```python
import pandas as pd
import networkx as nx

crypto_df = pd.read_csv("dataset/crypto.csv",sep=';')
tickers_df = pd.read_csv("dataset/ticker.csv",sep=';')
currency_df = pd.read_csv("dataset/currency.csv",sep=';')
```

```python
# Crear un grafo vaco
G = nx.Graph()

# Agregar nodos de criptomonedas y monedas fiduciarias al grafo
G.add_nodes_from(crypto_df['symbol'], bipartite=0)
G.add_nodes_from(currency_df['symbol'], bipartite=1)

# Agregar conexiones entre criptomonedas y monedas fiduciarias usando tickers
for _, row in tickers_df.iterrows():
    G.add_edge(row['symbol1'], row['symbol2'])

# Ahora G es un grafo bipartito donde las criptomonedas y las monedas fiduciarias estn conectadas segn los tickers

# Si deseas visualizar el grafo, puedes usar networkx o matplotlib
import matplotlib.pyplot as plt

# Posiciona los nodos usando el algoritmo bipartito de posicin de resorte
pos = nx.bipartite_layout(G, crypto_df['symbol'])

# Dibuja los nodos y las aristas
nx.draw(G, pos, with_labels=True, node_size=120, font_size=5)
plt.show()
```
